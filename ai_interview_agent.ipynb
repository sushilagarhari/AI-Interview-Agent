```python
import google.generativeai as genai
from google.colab import userdata

# API key setup
api_key = userdata.get('GEMINI_API_KEY')
genai.configure(api_key=api_key)

model = genai.GenerativeModel('gemini-2.5-flash')

# Interview questions bank
questions = [
    "What is the difference between supervised and unsupervised learning?",
    "Explain overfitting and how you can prevent it.",
    "What is the difference between a list and a tuple in Python?",
    "How does a Random Forest algorithm work?",
    "What is the purpose of the train-test split in machine learning?"
]

print("=" * 60)
print("AI INTERVIEW AGENT - Data Science Mock Interview")
print("=" * 60)
print(f"\nTotal Questions: {len(questions)}\n")
```

```python
scores = []
feedbacks = []

for i, question in enumerate(questions, 1):
    print(f"\n{'='*60}")
    print(f"Question {i}/{len(questions)}:")
    print(f"{question}")
    print(f"{'='*60}")

    user_answer = input("\nYour Answer: ")

    # AI se evaluation karwao
    prompt = f"""You are an expert Data Science interviewer evaluating a candidate's answer.

Question: {question}
Candidate's Answer: {user_answer}

Evaluate this answer and respond in this exact format:
Score: [give a score out of 10]
Feedback: [2-3 sentences of constructive feedback - what was good, what was missing]
"""

    response = model.generate_content(prompt)
    result = response.text

    print(f"\n--- AI Evaluation ---")
    print(result)

    feedbacks.append(result)

    # Score extract karne ki simple koshish
    try:
        import re
        score_line = [line for line in result.split('\n') if 'Score' in line][0]
        match = re.search(r'(\d+(\.\d+)?)\s*/\s*10', score_line)
        score = float(match.group(1))
        scores.append(score)
    except:
        scores.append(0)
print("\n\nInterview Complete!")
```

```python
print("="*60)
print("FINAL INTERVIEW SUMMARY")
print("="*60)

avg_score = sum(scores) / len(scores)
print(f"\nTotal Questions Answered: {len(questions)}")
print(f"Scores: {scores}")
print(f"Average Score: {avg_score:.1f}/10")

if avg_score >= 8:
    verdict = "Excellent! You're well-prepared for interviews."
elif avg_score >= 6:
    verdict = "Good performance! A bit more practice will make you interview-ready."
else:
    verdict = "Keep practicing — focus on core concepts and structured answers."

print(f"\nVerdict: {verdict}")
print("="*60)
```
